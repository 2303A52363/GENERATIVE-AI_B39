```python
import numpy as np

# Sigmoid Activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Linear Activation function (for output)
def linear(x):
    return x

def linear_derivative(x):
    return 1

# Training Data (Input: x1, x2, Output: y)
X_train = np.array([[0.2, 0.1], [0.3, 0.2], [0.4, 0.3], [0.7, 0.6], [0.8, 0.7], [0.9, 0.8]])
y_train = np.array([[0.3441], [0.3500], [0.3558], [0.3729], [0.3785], [0.3841]])

# Initialize parameters (weights and biases)
input_layer_size = 2
hidden_layer_size = 1
output_layer_size = 1

# Random initialization of weights and biases
np.random.seed(42)
W1 = np.random.randn(input_layer_size, hidden_layer_size)
b1 = np.random.randn(hidden_layer_size)
W2 = np.random.randn(hidden_layer_size, output_layer_size)
b2 = np.random.randn(output_layer_size)

# Learning Rate
learning_rate = 0.1
epochs = 10000  # Number of iterations

# Backpropagation with Gradient Descent
for epoch in range(epochs):
    # Forward Pass
    z1 = np.dot(X_train, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = linear(z2)

    # Compute the error (loss)
    loss = a2 - y_train
    mse = np.mean(loss**2)

    # Backward Pass (compute gradients)
    d_a2 = loss
    d_z2 = d_a2 * linear_derivative(a2)
    d_W2 = np.dot(a1.T, d_z2)
    d_b2 = np.sum(d_z2, axis=0, keepdims=True)

    d_a1 = np.dot(d_z2, W2.T)
    d_z1 = d_a1 * sigmoid_derivative(a1)
    d_W1 = np.dot(X_train.T, d_z1)
    d_b1 = np.sum(d_z1, axis=0, keepdims=True)

    # Update weights and biases
    W1 -= learning_rate * d_W1
    b1 -= learning_rate * np.squeeze(d_b1)
    W2 -= learning_rate * d_W2
    b2 -= learning_rate * np.squeeze(d_b2)

    # Every 1000 epochs, print the loss
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, MSE Loss: {mse:.6f}")

# Test Data (For MSE calculation and prediction)
X_test = np.array([[0.5, 0.4], [0.6, 0.5]])
y_test = np.array([[0.3615], [0.3672]])

# Predict the output on training and testing data
def predict(X):
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    return linear(z2)

# Calculate MSE on Test Data
y_pred_train = predict(X_train)
y_pred_test = predict(X_test)

train_mse = np.mean((y_pred_train - y_train)**2)
test_mse = np.mean((y_pred_test - y_test)**2)

print(f"Training MSE: {train_mse:.6f}")
print(f"Testing MSE: {test_mse:.6f}")

# Function to get user input and predict output
def user_input_predict():
    x1 = float(input("Enter x1: "))
    x2 = float(input("Enter x2: "))
    user_input = np.array([[x1, x2]])
    prediction = predict(user_input)
    print(f"Predicted output: {prediction[0][0]:.4f}")

# Uncomment the below line to take user input for prediction
user_input_predict()
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Training data
X_train = np.array([[0.2, 0.1], [0.3, 0.2], [0.4, 0.3], [0.7, 0.6], [0.8, 0.7], [0.9, 0.8]])
y_train = np.array([[0.3441], [0.3500], [0.3558], [0.3729], [0.3785], [0.3841]])

# Test data
X_test = np.array([[0.5, 0.4], [0.6, 0.5]])
y_test = np.array([[0.3615], [0.3672]])

# ANN architecture
input_neurons = 2
hidden_neurons = 3
output_neurons = 1
learning_rate = 0.1
epochs = 5000

# Initialize weights and biases
np.random.seed(42)
weights_input_hidden = np.random.rand(input_neurons, hidden_neurons)
bias_hidden = np.random.rand(1, hidden_neurons)
weights_hidden_output = np.random.rand(hidden_neurons, output_neurons)
bias_output = np.random.rand(1, output_neurons)

# Training loop
for epoch in range(epochs):
    # Forward propagation
    hidden_input = np.dot(X_train, weights_input_hidden) + bias_hidden
    hidden_output = sigmoid(hidden_input)
    final_input = np.dot(hidden_output, weights_hidden_output) + bias_output
    final_output = final_input  # Linear activation

    # Compute error
    error = y_train - final_output

    # Backpropagation
    d_output = error  # Derivative of linear activation is 1
    d_hidden = d_output.dot(weights_hidden_output.T) * sigmoid_derivative(hidden_output)

    # Update weights and biases
    weights_hidden_output += hidden_output.T.dot(d_output) * learning_rate
    bias_output += np.sum(d_output, axis=0, keepdims=True) * learning_rate
    weights_input_hidden += X_train.T.dot(d_hidden) * learning_rate
    bias_hidden += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

# Predictions and MSE Calculation
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Train error
hidden_output_train = sigmoid(np.dot(X_train, weights_input_hidden) + bias_hidden)
train_predictions = np.dot(hidden_output_train, weights_hidden_output) + bias_output
train_mse = mse(y_train, train_predictions)

# Test error
hidden_output_test = sigmoid(np.dot(X_test, weights_input_hidden) + bias_hidden)
test_predictions = np.dot(hidden_output_test, weights_hidden_output) + bias_output
test_mse = mse(y_test, test_predictions)

print(f"Training MSE: {train_mse:.6f}")
print(f"Testing MSE: {test_mse:.6f}")

# Predict function
def predict(x1, x2):
    x_input = np.array([[x1, x2]])
    hidden_output = sigmoid(np.dot(x_input, weights_input_hidden) + bias_hidden)
    output = np.dot(hidden_output, weights_hidden_output) + bias_output
    return output[0, 0]

# User input for prediction
x1, x2 = map(float, input("Enter x1 and x2: ").split())
predicted_output = predict(x1, x2)
print(f"Predicted Output: {predicted_output:.6f}")
```
